```python
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = None  # Will be generated automatically
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock


class Admin:
    def __init__(self):
        self.food_items = []
        self.current_food_id = 1

    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = FoodItem(name, quantity, price, discount, stock)
        food_item.food_id = self.current_food_id
        self.current_food_id += 1
        self.food_items.append(food_item)
        print("Food item added successfully!")

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for food_item in self.food_items:
            if food_item.food_id == food_id:
                food_item.name = name
                food_item.quantity = quantity
                food_item.price = price
                food_item.discount = discount
                food_item.stock = stock
                print("Food item edited successfully!")
                return
        print("Food item not found.")

    def view_food_items(self):
        print("Food items:")
        for food_item in self.food_items:
            print(f"Food ID: {food_item.food_id}")
            print(f"Name: {food_item.name}")
            print(f"Quantity: {food_item.quantity}")
            print(f"Price: {food_item.price}")
            print(f"Discount: {food_item.discount}")
            print(f"Stock: {food_item.stock}")
            print("----------")

    def remove_food_item(self, food_id):
        for food_item in self.food_items:
            if food_item.food_id == food_id:
                self.food_items.remove(food_item)
                print("Food item removed successfully!")
                return
        print("Food item not found.")

    def login(self):
        username = input("Enter admin username: ")
        password = input("Enter admin password: ")

        # Check if username and password match
        if username == "admin" and password == "admin123":
            print("Admin login successful!")
            return True
        else:
            print("Invalid username or password.")
            return False


class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.order_history = []

    def place_new_order(self, food_items):
        selected_items = []
        for item_number in food_items:
            if item_number <= len(admin.food_items):
                selected_items.append(admin.food_items[item_number - 1])
        if selected_items:
            print("Selected items:")
            for item in selected_items:
                print(f"Food ID: {item.food_id}")
                print(f"Name: {item.name}")
                print(f"Quantity: {item.quantity}")
                print(f"Price: {item.price}")
                print(f"Discount: {item.discount}")
                print(f"Stock: {item.stock}")
                print("----------")
            place_order = input("Place the order? (y/n): ")
            if place_order.lower() == 'y':
                self.order_history.append(selected_items)
                print("Order placed successfully!")
                # Decrease the stock of ordered items
                for item in selected_items:
                    item.stock -= 1
            else:
                print("Order canceled.")
        else:
            print("No valid items selected.")

    def view_order_history(self):
        if self.order_history:
            print("Order history:")
            for i, order in enumerate(self.order_history):
                print(f"Order {i+1}:")
                for item in order:
                    print(f"Food ID: {item.food_id}")
                    print(f"Name: {item.name}")
                    print(f"Quantity: {item.quantity}")
                    print(f"Price: {item.price}")
                    print(f"Discount: {item.discount}")
                    print(f"Stock: {item.stock}")
                    print("----------")
        else:
            print("No order history found.")

    def update_profile(self):
        new_full_name = input("Enter new full name: ")
        new_phone_number = input("Enter new phone number: ")
        new_email = input("Enter new email: ")
        new_address = input("Enter new address: ")
        new_password = input("Enter new password: ")

        self.full_name = new_full_name
        self.phone_number = new_phone_number
        self.email = new_email
        self.address = new_address
        self.password = new_password
        print("Profile updated successfully!")


admin = Admin()
user = None

while True:
    print("------------")
    print("1. Admin Login")
    print("2. User Login")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        if admin.login():
            while True:
                print("------------")
                print("Admin Menu:")
                print("1. Add New Food Item")
                print("2. Edit Food Item")
                print("3. View Food Items")
                print("4. Remove Food Item")
                print("5. Logout")
                admin_choice = input("Enter your choice: ")

                if admin_choice == "1":
                    name = input("Enter food item name: ")
                    quantity = input("Enter food item quantity: ")
                    price = float(input("Enter food item price: "))
                    discount = float(input("Enter food item discount: "))
                    stock = int(input("Enter food item stock: "))
                    admin.add_food_item(name, quantity, price, discount, stock)
                elif admin_choice == "2":
                    food_id = int(input("Enter food item ID to edit: "))
                    name = input("Enter new food item name: ")
                    quantity = input("Enter new food item quantity: ")
                    price = float(input("Enter new food item price: "))
                    discount = float(input("Enter new food item discount: "))
                    stock = int(input("Enter new food item stock: "))
                    admin.edit_food_item(food_id, name, quantity, price, discount, stock)
                elif admin_choice == "3":
                    admin.view_food_items()
                elif admin_choice == "4":
                    food_id = int(input("Enter food item ID to remove: "))
                    admin.remove_food_item(food_id)
                elif admin_choice == "5":
                    break
                else:
                    print("Invalid choice. Please try again.")

    elif choice == "2":
        print("------------")
        print("User Menu:")
        print("1. Register")
        print("2. Login")
        print("3. Exit")
        user_choice = input("Enter your choice: ")

        if user_choice == "1":
            full_name = input("Enter full name: ")
            phone_number = input("Enter phone number: ")
            email = input("Enter email: ")
            address = input("Enter address: ")
            password = input("Enter password: ")
            user = User(full_name, phone_number, email, address, password)
            print("User registration successful!")

        elif user_choice == "2":
            if user is None:
                print("User not registered.")
            else:
                email = input("Enter email: ")
                password = input("Enter password: ")
                if user.email == email and user.password == password:
                    print("User login successful!")
                    while True:
                        print("------------")
                        print("User Menu:")
                        print("1. Place New Order")
                        print("2. Order History")
                        print("3. Update Profile")
                        print("4. Logout")
                        user_option = input("Enter your choice: ")

                        if user_option == "1":
                            print("Available Food Items:")
                            admin.view_food_items()
                            selected_items = input("Enter the item numbers separated by commas: ")
                            selected_items = [int(item) for item in selected_items.split(",")]
                            user.place_new_order(selected_items)

                        elif user_option == "2":
                            user.view_order_history()

                        elif user_option == "3":
                            user.update_profile()

                        elif user_option == "4":
                            break



                        else:
                            print("Invalid choice. Please try again.")

                else:
                    print("Invalid email or password.")

        elif user_choice == "3":
            break

        else:
            print("Invalid choice. Please try again.")

    elif choice == "3":
        break

    else:
        print("Invalid choice. Please try again.")



```
